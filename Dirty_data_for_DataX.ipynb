```python
def impute_nans(df, amount_nan):
    cols_list = [col_num for col_num in range(df.shape[1])]
    rows_list = [row_num for row_num in range(df.shape[0])]
    for i in range(amount_nan):
        col_num = random.choice(cols_list)
        row_num = random.choice(rows_list)
        df.iloc[row_num, col_num] = np.nan
    return df
```

```python
def create_outliers(df, amount_outliers, multiplier = 3):
    cols_list = [col_num for col_num in range(df.shape[1])]
    rows_list = [row_num for row_num in range(df.shape[0])]
    count = 0
    for column in df.columns:
        if (df[column].dtype == 'float64' or
                df[column].dtype == 'int64'):
            count += 1
            break
    for i in range(amount_outliers):
        if count == 0:
            break
        if i%10 == 0:
            multiplier*= -1 
        while True:
            col_num = random.choice(cols_list)
            if (df.iloc[:, col_num].dtype == 'float64' or
                df.iloc[:, col_num].dtype == 'int64'):
                break
        col_name = df.columns[col_num]
        q1 = np.percentile(list(df.loc[df.loc[:,col_name].notnull(), col_name]),25)
        q3 = np.percentile(list(df.loc[df.loc[:,col_name].notnull(), col_name]),75)
        iqr = q3 - q1
        ceiling = q3 + 1.5*iqr
        row_num = random.choice(rows_list)
        if df.iloc[:, col_num].dtype == 'int64' and df.iloc[row_num, col_num] != np.nan:
            if np.sign(ceiling) == -1:
                df.iloc[row_num, col_num] =  int(ceiling*multiplier) 
            else: df.iloc[row_num, col_num] =  int(ceiling*multiplier) + random.randint(0, int(ceiling + 2))
        if df.iloc[row_num, col_num] != np.nan:
            df.iloc[row_num, col_num] =  ceiling*multiplier + random.random()
        if multiplier < 0:
            multiplier*= -1
    return df
```

```python
def impute_duplicates(df, amount_duplicates, max_rep):
    rep_num_list = [rep_num for rep_num in range(max_rep)]
    rows_list = [row_num for row_num in range(df.shape[0])]
    for i in range(amount_duplicates):
        row_num = random.choice(rows_list)
        rep_num = random.choice(rep_num_list)
        dup_row = df.iloc[[row_num]]
        for i in range(rep_num):
            df = pd.concat([df, dup_row], ignore_index=True)
    df = df.sample(frac=1).reset_index(drop=True)
    return df
```

```python
#Lets make the data dirty
if df.shape[0] < 5000:
    number = 80
elif  1000 < df.shape[0] < 20000:
    number = int(df.shape[0]*0.04)
elif  20000 < df.shape[0] < 100000:
    number = int(df.shape[0]*0.02)
elif  df.shape[0] >= 100000:
    number = 700
df = create_outliers(df, number, 100)
df = impute_nans(df, number)
df = impute_duplicates(df, int(number/2), 2)
```
