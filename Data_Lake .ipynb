```python
name_column = 'Full name'
submission_column = 'Last modified (submission)'
identifier = 'LD1'
def get_submission_date(df, name_column, submission_column, identifier):
    df = df.loc[df[submission_column] != '-']
    df= df.loc[:, [name_column, submission_column]]
    df.index = df[name_column]
    df= df.drop([name_column], axis = 1)
    df[submission_column] = df[submission_column].apply(lambda x: x.split(','))
    df[f'{identifier}. Submission Day'] = df[submission_column].apply(lambda x: x[0])
    df[f'{identifier}. Submission Date'] = df[submission_column].apply(lambda x: x[1])
    df[f'{identifier}. Submission Time'] = df[submission_column].apply(lambda x: x[2])
    df= df.drop([submission_column], axis = 1)
    return df
```

```python
#create new dataframe for learning diaries submission time
ld_sub_time_eng_1 = get_submission_date(grades_ld_eng_1, 'Full name', 'Last modified (submission)', 'LD1')
ld_sub_time_de_1 = get_submission_date(grades_ld_de_1, 'Full name', 'Last modified (submission)', 'LD1')
ld_sub_time_eng_2 = get_submission_date(grades_ld_eng_2, 'Vollständiger Name', 'Zuletzt geändert (Abgabe)', 'LD2')
ld_sub_time_de_2 = get_submission_date(grades_ld_de_2, 'Full name', 'Last modified (submission)', 'LD2')
```

```python
#Merging several dataframes to one with all participants
```

```python
def full_dataframe(column_names, rows, list_of_datasets):
    #create dictionary for empty dataframe
    dict_na = defaultdict(list)
    for name in column_names:
        dict_na[name] = [np.NaN for i in range(len(rows))]
    full_df = pd.DataFrame(dict_na, index = rows)
    #filling the dataset with answers of students:
    for df in list_of_datasets:
        #names of the students who submitted:
        names_submitted = list(df.index)
        for name in names_submitted:
            if name in names:
                full_df.loc[name,list(df.columns)] = df.loc[name,:]
    return full_df
```

```python
#full dataset with all submission times
full_df_sub = full_dataframe(sub_col_names_list, names, sub_df_list)
```

```python
#create column names to distinguish between weeks
def rename_columns(df, addition, number):
    rename_dict = {}
    for col_name in list(df.columns):
        rename_dict[col_name] = f'{addition}{number}. {col_name}'
    #preparing dictionaries to rename columns
    df = df.rename(columns = rename_dict, inplace = False)
    return df
```

```python
#rename column names:
ld_w_2 = rename_columns(ld_w_2, 'LD', 1)
ld_w_10 = rename_columns(ld_w_10, 'LD', 2)
```
